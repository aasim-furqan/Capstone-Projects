```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# To load the data
train = pd.read_csv('salary-prediction-for-job-postings/usjobs_train.csv')
print(train.shape)
train.head(5)
```

```python
test = pd.read_csv('salary-prediction-for-job-postings/usjobs_test.csv')
print(test.shape)
test.head(5)
```

```python
submission = pd.read_csv('salary-prediction-for-job-postings/usjobs_sample_submission.csv')
print(submission.shape)
submission.head(5)
```

```python
def rearrange_dataset(train_data, test_data):
    #1.train data
    skills_list = []
    for i in range(len(train_data)):
        zip_list = train_data['Skills'][i].replace('[','').replace(']','').replace("'",'').replace(" ",'').split(',')
        dict_temp = {}
        str_ = ''
        for j in range(len(zip_list)):
            dict_temp[zip_list[j]] = 1
            str_ += zip_list[j] + " "
        dict_temp['skills_clean'] = str_    
        skills_list.append(dict_temp)
        
    train_data = pd.concat([train_data,pd.DataFrame(skills_list).fillna(0)],axis =1)  
    
    
    #2.test data
    skills_list = []
    for i in range(len(test_data)):
        zip_list = test_data['Skills'][i].replace('[','').replace(']','').replace("'",'').replace(" ",'').split(',')
        dict_temp = {}
        str_ = ''
        for j in range(len(zip_list)):
            dict_temp[zip_list[j]] = 1
            str_ += zip_list[j] + " "
        dict_temp['skills_clean'] = str_        
        skills_list.append(dict_temp)
        
    test_data = pd.concat([test_data,pd.DataFrame(skills_list).fillna(0)],axis =1)      
    
    return train_data, test_data
```

```python
%%time
train_data_arr, test_data_arr = rearrange_dataset(train_data= train, test_data= test)
```

```python
train_data_arr[['Job','Company','skills_clean','ArtificialIntelligence', 'MachineLearning', 'Python', 'TensorFlow',
       'Excel', 'Spark', 'PhD', 'AWS', 'C++', 'DeepLearning', 'Java', 'Master',
       'Office', 'Docker', 'Word', 'Azure', 'Hadoop', 'Airflow', 'MBA', 'CPA',
       'Director_Score', 'Snowflake', 'Databricks', 'SQL', '', 'Company_Score',
       'GoogleCloud', 'SciKit', 'Reviews', 'Pandas', 'ChatGPT', 'Agile', 'R',
       'NeuralNetwork', 'Jupyter', 'PowerPoint', 'Spanish', 'NumPy', 'React',
       'GitHub', 'Looker', 'English', 'Access', 'Matplotlib', 'Seaborn', 'C#',
       'Rust', 'VBA', 'Bachelor']].isna().sum()
```

```python
train_data_arr['Director_Score'].fillna(np.mean(train_data_arr['Director_Score']), inplace= True)
train_data_arr['Company_Score'].fillna(np.mean(train_data_arr['Company_Score']), inplace= True)
train_data_arr['Reviews'].fillna(np.mean(train_data_arr['Reviews']), inplace= True)
```

```python
# To tarin 
X_train, X_test, y_train, y_test = train_test_split(train_data_arr[['ArtificialIntelligence', 'MachineLearning', 'Python', 'TensorFlow',
       'Excel', 'Spark', 'PhD', 'AWS', 'C++', 'DeepLearning', 'Java', 'Master',
       'Office', 'Docker', 'Word', 'Azure', 'Hadoop', 'Airflow', 'MBA', 'CPA',
       'Director_Score', 'Snowflake', 'Databricks', 'SQL', '', 'Company_Score',
       'GoogleCloud', 'SciKit', 'Reviews']],
                                                   train_data_arr['Mean_Salary'],
                                                   test_size = 0.2)
```

```python
clf = RandomForestRegressor(n_estimators= 500,
                            max_depth = 30,
                            min_samples_leaf = 25,
                           n_jobs= 5)

clf.fit(X_train,
       y_train)
```

```python
mean_absolute_error(y_test,clf.predict(X_test))
```

```python
test_data_arr['Director_Score'].fillna(np.mean(train_data_arr['Director_Score']), inplace= True)
test_data_arr['Company_Score'].fillna(np.mean(train_data_arr['Company_Score']), inplace= True)
test_data_arr['Reviews'].fillna(np.mean(train_data_arr['Reviews']), inplace= True)
```

```python
test_df = test_data_arr[['ID','ArtificialIntelligence', 'MachineLearning', 'Python', 'TensorFlow',
       'Excel', 'Spark', 'PhD', 'AWS', 'C++', 'DeepLearning', 'Java', 'Master',
       'Office', 'Docker', 'Word', 'Azure', 'Hadoop', 'Airflow', 'MBA', 'CPA',
       'Director_Score', 'Snowflake', 'Databricks', 'SQL', '', 'Company_Score',
       'GoogleCloud', 'SciKit', 'Reviews']]
```
